## Join and Merge

```python
Employees = pd.DataFrame({'ID': [1, 2, 3, 4, 5],
    'Name': ['Sara', 'Katie', 'David', 'Hazal', 'Georgie'],
    'Address': [42, 38, 36, 41, 35],
    'Salary': [33000, 45000, 65000, 95000, 35000]
})
Employees
```

```python
Department = pd.DataFrame({
    'ID' :[1,3,5,7],
     'Course':['Sales','HR','Marketing','Finance']
})
Department
```

```python
#inner join=>matching rows
inner_join_df=pd.merge(Employees, Department, on="ID", how="inner")
inner_join_df
```

```python
#outer join=>all rows form the two dataframe
outer_join_df=pd.merge(Employees, Department, on="ID", how="outer")
outer_join_df
```

```python
#left join=> all from first and only matching rows
left_join_df=pd.merge(Employees, Department, on="ID", how="left")
print(left_join_df)
```

```python
#right join=> all from second and only matching from first
right_join_df=pd.merge(Employees, Department, on="ID", how="right")
print(right_join_df)
```

```python
df_con=pd.concat([Employees,Department], axis=1)
df_con
```

```python
Marketing = pd.DataFrame({
    'ID' :[1,2,3,7],
     'Category':['Snacks','Bars','Cookies','bites']
})
Marketing
```

```python
df_con=pd.concat([Employees,Department, Marketing], axis=1)
df_con
```

```python
left_join_df=pd.merge(pd.merge(Employees, Department, on ="ID",how="left"),Marketing, on="ID",how="left")#leftjoin for more than 2 data frame
left_join_df
```

```python
df_con=pd.concat([Employees,Department, Marketing], axis=1)#concat for more than 2 data frame
df_con
```

```python
df_con=pd.concat([Employees,Department], axis=1)#concat for 2 data frame
df_con
```

# Map()

```python
mylist=[2,4,6,8,10]
def funcx(x):
  return x**x
newlist=[]
for x in mylist:
  newlist.append(funcx(x))
print(newlist)
```

```python
mylist=[2,4,6,8,10]
def funcx(x):
  return x**x
print(list(map(funcx,mylist)))
```
